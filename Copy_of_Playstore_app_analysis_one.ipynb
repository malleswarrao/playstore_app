```python
gk = new_cf.groupby('Category').sum()
pk = new_cf.groupby('Category').size()
#avg_rating=gk.loc['BEAUTY']/pk['BEAUTY']
for i in gk.iterrows():
  avg_rating=float(gk.loc[i[0]]/pk[i[0]])
  for k in new_cf.iterrows():
    if(k[1][1]==i[0]):
      if(str(k[1][2])=='nan'):
        new_cf.iloc[k[0],2]=avg_rating
```

```python
new_cf.tail(20)
```

```python
new_cf.dropna(inplace=True)
```

```python
def modify_me(input_x):
  input_x=int(input_x.replace(",",""))
  #print(input_x)
  return input_x

new_cf['Installs'] = new_cf['Installs'].apply(lambda x:modify_me(x[:-1]))
#df.apply(lambda x: func(x['col1'],x['col2']),axis=1)
new_cf.head(20)
```

```python
new_cf.tail(30)
```

```python
def modify_size(input_x,z):
  if(z=='M'):
    kt = int(float(input_x[:-1]))*1000
  elif(z=='k'):
    kt = int(float(input_x[:-1]))
  else:
    kt = input_x
  return kt

new_cf['Size'] = new_cf['Size'].apply(lambda x:modify_size(x,x[-1]))
```
