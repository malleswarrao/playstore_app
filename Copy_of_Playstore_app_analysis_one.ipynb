```python
gk = new_cf.groupby('Category').sum()
pk = new_cf.groupby('Category').size()
#avg_rating=gk.loc['BEAUTY']/pk['BEAUTY']
for i in gk.iterrows():
  avg_rating=float(gk.loc[i[0]]/pk[i[0]])
  for k in new_cf.iterrows():
    if(k[1][1]==i[0]):
      if(str(k[1][2])=='nan'):
        new_cf.iloc[k[0],2]=avg_rating
```

```python
new_cf.dropna(inplace=True)
```

```python
def modify_me(input_x):
  input_x=int(input_x.replace(",",""))
  #print(input_x)
  return input_x

new_cf['Installs'] = new_cf['Installs'].apply(lambda x:modify_me(x[:-1]))
#df.apply(lambda x: func(x['col1'],x['col2']),axis=1)
new_cf.head(20)
```

```python
new_cf.tail(30)
```

```python
def modify_size(input_x,z):
  if(z=='M'):
    kt = int(float(input_x[:-1]))*1000
  elif(z=='k'):
    kt = int(float(input_x[:-1]))
  else:
    kt = input_x
  return kt

new_cf['Size'] = new_cf['Size'].apply(lambda x:modify_size(x,x[-1]))
```

```python
new_cf.drop_duplicates(subset = None, keep = 'first')
```

```python
new_cf['Reviews'] = new_cf['Reviews'].astype(float)
new_cf['Installs'] = new_cf['Installs'].astype(float)
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def insights_func(new_df_k,ki):
  size_vs_installs = new_df_k[['Size','Installs']]
  size_vs_installs = size_vs_installs.loc[~size_vs_installs['Size'].isin(['Varies with device'])]
  size_vs_installs.set_index('Size',inplace=True)
  size_vs_installs.sort_index(inplace=True)
  plt.figure()
  plt.title(ki)
  plt.xlabel('size') 
  plt.ylabel('Installs')
  plt.plot(size_vs_installs)
  plt.show()
  #insights for free vs paid installs
  paid_vs_free = new_df_k.groupby(['Type'])['Installs'].sum()
  paid_vs_free.plot(kind='bar')
  plt.title(ki)
  plt.show()
  #insights for content rating vs installs
  content_vs_installs = new_df_k.groupby(['Content Rating'])['Installs'].sum()
  content_vs_installs.plot(kind = 'bar')
  plt.title(ki)
  plt.show()
  #insights for rating vs installs
  Rating_vs_installs = new_df_k[['Rating','Installs']]
  # size_vs_installs = size_vs_installs.loc[~size_vs_installs['Size'].isin(['Varies with device'])]
  Rating_vs_installs.set_index('Rating',inplace=True)
  Rating_vs_installs.sort_index(inplace=True)
  plt.figure()
  plt.title(ki)
  plt.xlabel('Rating') 
  plt.ylabel('Installs')
  plt.plot(Rating_vs_installs)
  plt.show()
  #insights for version vs installs
  version_vs_installs = new_df_k.groupby(['Android Ver'])['Installs'].sum()
  version_vs_installs.plot(kind = 'bar')
  plt.title(ki)
  plt.show()
  #insights for lastupdated vs installs
  new_tf=new_df_k
  new_tf['DATE']=new_tf['Last Updated'].str[-4:]
  updates_vs_installs = new_tf.groupby(['DATE'])['Installs'].sum()
  updates_vs_installs.plot(kind = 'bar')
  plt.title(ki)
  plt.show()

```

```python
kfd = (new_cf[['Category']])
tft =  kfd.values.tolist()
kft = []
for i in tft:
  kft.append(i[0])
kft = list(set(kft))
for i in kft:
  new_df = new_cf[new_cf['Category'] == i]
  insights_func(new_df,i)
```
